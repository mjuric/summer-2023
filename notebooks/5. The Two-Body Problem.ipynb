# 5. The Two-Body Problem

Previously, we've shown how to numerically solve the equations of motion of a body experiencing the force of gravity:

$$ \vec{F} = - G \frac{M m}{r^2} \hat{r} $$

But we've also seen the solution is (computationally) costly, if we want it to be precise. Also, the numerical solution doesn't provide much **insight** -- the ability to intuitively see, by just by looking at the result (the equation), what would happen if the initial conditions were slightly different, or the parameters of the problem changed (e.g. the masses). Both of these -- but especially the latter -- are good reasons to try to solve this problem:

$$ \ddot{\vec{r}} = \frac{\vec{F}_{\rm gravity}}{m} = - G \frac{M}{r^2} \hat{r} $$

analytically.

### Note: About Conservative Forces

There's one place in the handout that requires a little bit more discussion and explanation:



A few potential points of confusion here:

**First:** I think it's clearer to think (you'll soon see why) of the integrals above as _definite_ integrals, i.e., having specified _limits_ of integration. So in the first equation above, on the left-hand side we'd be integrating form some initial time $t_0$ to some final time $t_1$, while on the right-hand side we'd be integrating from some intiial position $\vec{r_0}$ to the final position $\vec{r_1}$.

If we did that, then the third equation would have the form:

$$ \left[ \frac{1}{2} m \dot{r}^2 \right]_{t_0}^{t_1} = \int_{\vec{r_0}}^\vec{r_1} \vec{F}(\vec{r}) \cdot d\vec{r} $$

We now need to figure out how to compute the integral on the right hands side...

**Next:** The manipulation of integrals -- especially changing the variable of integration from $t$ to $\vec{r}$ is quite clever -- maybe too clever?... E.g., while we understand what integrating over $t$ means, what does it even mean to integrate over $\vec{r}$? To understand this, go back to our definition of integrals as [Riemann sums](https://en.wikipedia.org/wiki/Riemann_sum): the result is a sum of the value of the function, times the (small) interval of a variable being integrated over, from some beginning to the end -- i.e., the proverbial "area under the curve".

A function (curve) with one variable, like $f(t)$ lives in two-dimensional space (i.e., can be sketched on an y-vs.-x graph. When we swap out the 1-d parameter like $t$ for a two-dimensional $\vec{r}(t)$, this function (curve) starts living in a three-dimensional space. To sketch it, we plot a line in 3D, with the value $f(x, y)$ being the height -- the $z$ axis. But we can still compute the area under that curve -- and that is what these integrals do.

Let's develop some analogies: Imagine our curve as a [rollercoaster](https://www.pennlive.com/resizer/Tflz__b640fcoq4RsfgXXCQdBew=/1280x0/smart/cloudfront-us-east-1.images.arcpublishing.com/advancelocal/MIRQLBADBVDGPFJRVN52BVHVFE.JPG). You can specify where you are on a rollercoaster in multiple equivalent ways. One is with a 1-d parameter such as "distance travelled from the start". Or the _time_ since the start of the ride (assuming all rides always unfold in exactly the same way). But you can also parametrize it by the 2D GPS coordnates (longitude, lattitude) of where you are on the roller coaster. And (and this is the important part) -- no matter how you've parametrized it, **you can still compute the area under the roller coaster, along the path the train track takes**.

This is what the substitution of variables above means. When we write:

$$ \int_{t_0}^{t_1} \vec{F}(\vec{r}) \cdot \vec{r} dt = \int_{\vec{r(t_0)}}^{\vec{r(t_1)}} \vec{F}(\vec{r}) \cdot d\vec{r} $$

we mean that the area under the curve of the function $\vec{F}(\vec{r}) \cdot \vec{r}$, calculated from $t_0$ to $t_1$ is the same as the area under the curve of $\vec{F}(\vec{r}) \cdot d\vec{r} $ ***calculated along the path the body would take*** while traveling under the influence of these same forces from $t_0$ to $t_1$. This bolded part is _important_: this integral **must** be summed up (in the Reimann sum sense) along not just any path, but the path the body would actually take. We must take a ride along the *same* roller-coaster.

**Then:** This now leaves us with a practical challenge of how to compute the integral:

$$ \int_{\vec{r(t_0)}}^{\vec{r(t_1)}} \vec{F}(\vec{r}) \cdot d\vec{r} $$

along the trajectory the body would take. But the reson we're doing this is in the first place is to _derive_ the equation for that trajectory (which happens a few pages later). The writeup plays a bit fast-and-loose with these issues, basically glossing over the problem by invoking "conservative forces". But this requires some explanation.

The reason we can get away with not knowing the exact path, and yet still evaluate the integral above, is that the path taken **doesn't matter** -- the integral will always be the same. It's easy to prove rigorously for this particular example of a **central force**, i.e., one where $\vec{F}(\vec{r}) = F(r)\cdot\hat{r}$:

$$ \int_{\vec{r_0}}^{\vec{r_1}} \vec{F}(\vec{r}) \cdot d\vec{r}  = 
   \int_{\vec{r_0}}^{\vec{r_1}} \left(F(r) \cdot \hat{r}\right) \cdot \left(dr \cdot \hat{r}\right) = 
   \int_{r_0}^{r_1} F(r) \cdot dr
$$

where we've _named_ the solution to the integral $\int_{\infty}^{R} F(r) \cdot dr$ as $V(R)$. What happened here was that since the force is *central*, the 2D integral along a path in $(r, \theta)$ plane collapsed down to a 1D integral along $r$. For a function of one variable, there's only one way to get from $r_0$ to $r_1$ and one unique solution -- it's just an area under the curve simple curve. Therefore, for any 2D path, the solution will be the same and depend only on $r_0$ and $r_1$, the distance from the center.

Note we could write that last integral as:

$$
   \int_{r_0}^{r_1} F(r) \cdot dr = 
       \int_{r_0}^{\infty} F(r) \cdot dr + \int_{\infty}^{r_1} F(r) \cdot dr
$$

This is just simple symbolic manipulation with boundaries of integration, that works for any integral. But the _physical interpretation_ is that of the *work* to move the body from $r_0$ to infinity, and then back from infinity to $r_1$. We give this integral the name *potential energy*, and with a bit more manipulation get the final expression:

If we **define**:

$$ V(r_0) = - \int_{r_0}^{\infty} F(r) \cdot dr $$

we get:

$$
   \int_{r_0}^{r_1} F(r) \cdot dr = V(r_1) - V(r_0)
$$

**Finally:** we can write our equation as:

$$ \left[ \frac{1}{2} m \dot{r}^2 \right]_{t_0}^{t_1} = V(r_1) - V(r_0) $$

and move the terms around to have all the final values on the left, and initial conditions on the right:

$$ \frac{1}{2} m \dot{r}(t_1)^2  + V(r_1) = \frac{1}{2} m \dot{r}(t_0)^2 + V(r_0) $$

The expression on the right is just some function of initial conditions, and therefore a **constant**. That means that the expression on the left will also always be a constant, for any time $t_1$ and the position of the body $r_1$ at that new time. That expression is just the **total energy**, and this just proves it's conserved.

PS: The introduction of the gradient $\nabla$, "potential", and "conservative forces" may seem a bit out of the blue (and it is, if you've had no introduction to them before). These generalize the above concepts to more than just central forces.
